```python
import pandas as pd

# 1. Charger le dataset original
df = pd.read_csv("Crimes_-_2001_to_Present.csv")

# 2. Convertir la colonne 'Date' en datetime
df['Date'] = pd.to_datetime(df['Date'])

# 3. Filtrer les données à partir de 2015
df_2015_now = df[df['Date'] >= "2015-01-01"]

# 4. Exporter les données filtrées dans un nouveau CSV
df_2015_now.to_csv("crimes_chicago_2015_2025.csv", index=False)

print("Extraction terminée !")
print("Nombre de lignes :", df_2015_now.shape[0])
```
